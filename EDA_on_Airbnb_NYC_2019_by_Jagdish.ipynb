## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
#Create a copy of the current dataset and assigning to df
df=airbnb_df.copy()
````

**Replacing the null values with appropriate values**

````python
#cleaning the data and replacing the null values
df['name'].replace(np.nan, "Other Hotel", inplace =True)

df['host_name'].replace(np.nan, 'other', inplace= True)

df['last_review'].replace(np.nan, 'Not Reviewed', inplace = True)

df['reviews_per_month'].replace(np.nan, '0', inplace= True)
````

````python
df.info()
````

````python
df.drop(['latitude','longitude'],axis=1,inplace=True) #since the value has no significancce in analysis.
````

````python
#Top 10 host_name
top_10_host=df.groupby('host_name',as_index=False).agg({'id':'count'}).rename(columns={'id':'host_count'}).sort_values('host_count',ascending=False).head(10).reset_index()
top_10_host
````

1. Host name is the name of the host who listed the hotel in the airbnb.

2. It looks like the person Michael has the largest booking under his name with 417 bookings

3. David is the host name with 403 bookings.

````python
neighbourhood_group_df=df.groupby('neighbourhood_group',as_index=False).agg({'id':'count'}).rename(columns={'id':'neighbourhood_group_count'}).sort_values('neighbourhood_group_count',ascending=False).reset_index()
neighbourhood_group_df.drop('index',axis=1,inplace=True)
neighbourhood_group_df
````

````python
#top 10 host_id
top_10_host_id=df.groupby('host_id',as_index=False).agg({'id':'count'}).rename(columns={'id':'booking_count'}).sort_values('booking_count',ascending=False).head(10).reset_index()
top_10_host_id.drop('index',axis=1,inplace=True)
top_10_host_id
````

````python
# diffrent room type and there count
room_type_df=df.groupby('room_type',as_index=False).agg({'id':'count'}).rename(columns={'id':'count_of_rooms'}).sort_values('count_of_rooms',ascending=False).reset_index()
room_type_df.drop('index',axis=1,inplace=True)
room_type_df
````

````python
#avg prices of differnt type of room
avg_type_df=df.groupby(['room_type'],as_index=False).agg({'price':'mean'}).rename(columns={'price':'average_price'}).sort_values('average_price',ascending=False).reset_index()
avg_type_df.drop('index',axis=1,inplace=True)
avg_type_df
````

````python

neighborhood_reviews = df.groupby('neighbourhood_group', as_index=False).agg({'number_of_reviews': 'sum'}).sort_values('number_of_reviews',ascending=False)
neighborhood_reviews
````

````python
room_type_reviews = df.groupby('room_type', as_index=False).agg({'number_of_reviews': 'sum'})
room_type_reviews
````

````python
#removing the outliers from the price column and creating a new filtered dataframe
# Calculate the first quartile (Q1) and third quartile (Q3)
Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)

# Calculate the interquartile range (IQR)
IQR = Q3 - Q1

# Define the lower and upper bounds to identify outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame to remove outliers based on the 'price' column
filtered_df = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]
filtered_df.describe()
````
