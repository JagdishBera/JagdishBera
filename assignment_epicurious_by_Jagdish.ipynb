### Data Wrangling Code

```python
# Write your code to make your dataset analysis ready.
#Create a copy of the current dataset and assigning to df
df=epicurious_df.copy()
```

```python
# Dropping duplicate rows
df.drop_duplicates(inplace=True)
# Dataset Duplicate Value Count
len(df[df.duplicated()])
```

**Replacing the null values with appropriate values**

```python
# Drop rows with missing values in 'protein', 'fat', or 'calories'
df.dropna(subset=['protein', 'fat', 'calories'], inplace=True)
```

```python
# Missing Values/Null Values Count after removing the value
print(df.isnull().sum())
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
#removing the outliers from the calories column and creating a new filtered dataframe
# Calculate the first quartile (Q1) and third quartile (Q3)
Q1 = df['calories'].quantile(0.25)
Q3 = df['calories'].quantile(0.75)

# Calculate the interquartile range (IQR)
IQR = Q3 - Q1

# Define the lower and upper bounds to identify outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame to remove outliers based on the 'calories' column
filtered_df = df[(df['calories'] >= lower_bound) & (df['calories'] <= upper_bound)]
filtered_df.describe()
```

```python
#removing the outliers from the sodium column and creating a new filtered dataframe
# Calculate the first quartile (Q1) and third quartile (Q3)
Q1 = filtered_df['sodium'].quantile(0.25)
Q3 = filtered_df['sodium'].quantile(0.75)

# Calculate the interquartile range (IQR)
IQR = Q3 - Q1

# Define the lower and upper bounds to identify outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the DataFrame to remove outliers based on the 'sodium' column
r_df = filtered_df[(filtered_df['sodium'] >= lower_bound) & (filtered_df['sodium'] <= upper_bound)]
r_df.describe()
```

# **Data Vizualization**

```python

dairy_free_count = len(r_df[r_df['dairy free'] == 1.0])
dairy_count = len(r_df[r_df['dairy free'] == 0.0])

labels = ['lactose-Free', 'lactose']
sizes = [dairy_free_count, dairy_count]
colors = ['#FFFFCC', '#FFFF00']  # Cream and Yellow colors
explode = (0.1, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')  # equal ratio
plt.title('Proportion of lactose-Free and lactose Dishes')
plt.show()
```

```python
veg_count = len(r_df[r_df['vegetarian'] == 1.0])
nonveg_count = len(r_df[r_df['vegetarian'] == 0.0])

labels = ['Vegetarian', 'Non-Vegetarian']
sizes = [veg_count, nonveg_count]
colors = ['green', 'darkred']  # Green and Dark Red colors
explode = (0.1, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')  # equal ratio
plt.title('Proportion of Vege and Non-Veg Dishes')
plt.show()
```

```python
# Count occurrences of breakfast, lunch, and dinner
breakfast_count = len(r_df[r_df['breakfast'] == 1])
lunch_count = len(r_df[r_df['lunch'] == 1])
dinner_count = len(r_df[r_df['dinner'] == 1])


# Count occurrences of breakfast, lunch, and dinner
meals = ['Breakfast', 'Lunch', 'Dinner']
counts = [breakfast_count, lunch_count, dinner_count]

# Plotting the graph with percentage values
plt.bar(meals, counts, color=['lightblue', 'blue', 'darkblue'])
plt.xlabel('Meals')
plt.ylabel('Count')
plt.title('Count of Different Meals')

# Adding percentage values on top of the bars
total = sum(counts)
for i, count in enumerate(counts):
    percentage = count / total * 100
    plt.text(i, count, f'{percentage:.1f}%', ha='center')

plt.show()
```

```python
# Grouping by 'rating' column and counting occurrences
rating_counts = r_df['rating'].value_counts()

# Plotting the graph
rating_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Count of Ratings')
plt.show()
```

```python
# Sort the DataFrame by the 'protein' column in descending order and fetch the top 10 rows
top_protein_dishes = r_df.nlargest(10, 'protein')[['title', 'protein']].reset_index()
top_protein_dishes
```

```python
top_rated_dishes = r_df.nlargest(5, 'rating')[['title', 'rating']]
top_rated_dishes
```

```python
# Count occurrences of specific columns
meals_22_min = len(r_df[r_df['22-minute meals'] == 1])
recipes_3_ing = len(r_df[r_df['3-ingredient recipes'] == 1])
prep_required = len(r_df[r_df['advance prep required'] == 1])

# Storing the values
categories = ['22-minute meals', '3-ingredient recipes', 'Advance prep required']
counts = [meals_22_min, recipes_3_ing, prep_required]

# Plotting the horizontal bar graph with value labels above each bar
plt.barh(categories, counts, color='skyblue')
plt.xlabel('Count')
plt.title('type of Recipe by time')

# Adding value labels above each bar
for index, value in enumerate(counts):
    plt.text(value, index, str(value))

plt.show()
```

```python
# Calculating the correlation matrix
correlation_matrix = r_df[['calories', 'fat', 'protein', 'sodium']].corr()

# Creating the heatmap
plt.figure(figsize=(6, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap="Reds")

# Setting the title
plt.title('Correlation Heatmap of Calories, Fat, Protein, and Sodium')

# Displaying the heatmap
plt.show()
```
